```python
from bio_embeddings.embed import ProtTransBertBFDEmbedder
from Bio import SeqIO
```

```python
seqs = []
for record in SeqIO.parse("data_files/deeploc_swissprot_clipped1k.fasta", "fasta"):
    seqs.append(record)
```

```python
seqs
```

```python
embs = embedder.embed_many([str(s.seq) for s in seqs])

# `embed_many` returns a generator.
# We want to keep both RAW embeddings and reduced embeddings in memory.
# To do so, we simply turn the generator into a list!
# (this will start embedding the sequences!)

embs = list(embs)
```

```python
reduced_embs = [embedder.reduce_per_protein(e) for e in embs]
```

```python
for (per_amino_acid, per_protein) in zip(embs, reduced_embs):
    print(per_amino_acid.shape, per_protein.shape)
```
